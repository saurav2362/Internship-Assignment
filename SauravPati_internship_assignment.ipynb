## Reviewing the article "Why is the Tower of God show so powerful?"

Let's imagine the article “Why is the Tower of God Show So Popular?” has user data 
such as page views, average time spent, and bounce rates.

Importing the required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {'Days' : ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'],'Page Views' : [750,1000,1100,1250,1300,1450,1700]}
df = pd.DataFrame(data)
df
```

Let's draw a bar plot to compare the page views of different days in a week.

```python
sns.set(style = 'whitegrid')
plt.figure(figsize = (7,4))
sns.barplot(x = 'Days',y = 'Page Views',data = df, width = 0.6)
plt.title('Trend of page views of Tower of God article')
plt.xlabel('Days')
plt.ylabel('Page views')
plt.show()
```

The page views are consistently increasing. It shows that the article is of broad interest.

```python
data1 = {'Avg Time Spent' : ['Less than 30 sec', '30 sec to 1 min', 'More than 1 min'],'Percentage' : [15,35,50]} 
df1 = pd.DataFrame(data1)
df1
```

```python
plt.pie(df1['Percentage'] ,labels = df1['Avg Time Spent'], autopct = '%0.1f%%',shadow = True,startangle = 90)
plt.title('Average time spent on Tower of God article')

plt.show()
```

Small percentage of users spend less time in the article as compared to the users who spend more. This shows that the article is enagaging and people are hooked to the article.

```python
data2 = {'Weeks' : ['Week 1', 'Week 2', 'Week 3', 'Week 4'], 'Bounce Rates (%)' : [65,55,50,30]}
df2 = pd.DataFrame(data2)
df2
```

```python
plt.plot(df2['Weeks'],df2['Bounce Rates (%)'],marker = 'o')
plt.title('Weekly Bounce rates of Tower of God article')
plt.xlabel('Weeks')
plt.ylabel('Bounce rate (%)')
plt.show()
```

The bounce rates keeps on decreasing. This suggests that the users find the articles more engaging and thus they are likely to visit other pages as well. 

Since it has high page views, more time spent in the article and also low bounce rates, it indicates that the users find this article more compelling enough to stay.

###  User segmentation based on demographics or behavior

#### **Demographics based segmentation**

We can seggregate content for different age groups on the basis of what they actually want to see.
For example:
- **Teenagers** (age 13 to 18) seek more action packed animes with fascinating visuals. They are more interested in cool moments and plot twists. In order to engage them with articles we can use titles such as "Top 5 best fights in Solo Leveling" or "Facts about Solo Leveling that will blow your mind".
- **Young Adults** (age 19 to 30) seek character arcs, storyline, the thrill, etc. To captivate them we can use titles such as "Is Jeju Island arc the best arc in Solo Leveling?". Diving deep into the arcs in addition with cool pictures will make the Young Adults craving for more.
- **Adults** (age 30 or more) appreciate the cultural touch, world building and literary themes such as heroism, sacrifice for the greater good. They like metaphor or allegory. They would want more sophisticated articles comparing Solo Leveling with the real world. Use titles such as "How Solo Leveling Reflects Modern Hero Archetypes in Korean Manhwa". We can also try using interviews or reviews from industry experts and discussion forums.

We can seggregate articles for different regions too. For example:
- **North America/Europe**: Users from this region might be familiar with the popular mangas such as Attack on Titan or Bleach or Naruto. Try drawing parallels with this type of animes for broader appeal.
- **Asia**: Users from this region are likely familiar with the cultural nuances. Try pointing out the different cultural meanings behind certain references or cultural aspects related to Korean Manhwa.

#### **Behavioural Segmentation**

Tailoring content for new and returning users.
- **New Users**: These readers might be exploring Solo Leveling for the first time and so we need to create introductory articles with titles such as "A Beginner's Guide to Solo Leveling" or "What makes Solo Leveling so epic".
- **Returning readers**: These readers are more likely to be long time fans. They might get indulge in detailed analyses and discussions. Use titles such as "10 Best Arcs in Solo Leveling, Ranked", "Hidden facts about Solo Leveling that will blow your mind". Use polls or comment forums to encourage community engagement.


Tailoring content for low and high engagement users.
- **Low Engagement Users**: They are more likely to skim through the whole article. We would need more visually pleasing articles and crisp contents. Use titles such as "Top 3 moments in Solo Leveling" or "Top 3 character arcs in Solo Leveling".
- **High Engagement Users**: These are readers who spend more time reading articles. They are more likely to indulge in long form content. We can try creating arc by arc comparison, character analyses followed by community engagement through comment forums. Use titles such as " From the Double Dungeon to the Monarch Wars, let's see how far Solo Leveling has come".
